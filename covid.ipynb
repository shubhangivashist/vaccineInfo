```python
import json
import requests
from datetime import date
import re
today = date.today()

def ValidPincode(pincode):
    regex="^[1-9]{1}[0-9]{2}\\s{0,1}[0-9]{3}$"
    p=re.compile(regex)
    #if (pincode == ''):
     #   return False
    m = re.match(p, pincode);
    if m is None:
        return False
    else:
        return True

def GetStateId(state_name):
    url="https://cdn-api.co-vin.in/api/v2/admin/location/states"
    header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:88.0) Gecko/20100101 Firefox/88.0"}
    response=requests.get(url,headers=header)
    var=response.json()
    par=0
    a=var['states']
    state=[]
    for i in a:
        state.append(i['state_name'].lower())
    if state_name in state:
        for i in a:
            if i['state_name'].lower()==state_name:
                return (i['state_id'])
    else:
        return 0

def GetDistrictId(district_name,state_id):
    url2="https://cdn-api.co-vin.in/api/v2/admin/location/districts/"+state_id
    header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:88.0) Gecko/20100101 Firefox/88.0"}
    response=requests.get(url2,headers=header)
    var=response.json()
   
    district=[]
    a=var['districts']
    for i in a:
        district.append(i['district_name'].lower())
    
    if district_name in district:
        for i in a:
            if i['district_name'].lower()==district_name:
                return(i['district_id'])
    else:
        return 0
                
def Availability(pincode,district_id):
    date=today.strftime("%d-%m-%Y")
   
    if district_id==0:
        url="https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/findByPin"
        params={"pincode":pincode,"date":date}
        header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:88.0) Gecko/20100101 Firefox/88.0"}
        response=requests.get(url,params,headers=header)
        var=response.json()
        a=var['sessions']
        for i in a:
            if i['available_capacity']!=0:
                print ("Name:"+i['name']+" "+"Vaccine:"+i['vaccine'])
            
    else:
        url="https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/findByDistrict"
        params={"district_id":district_id,"date":date}
        header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:88.0) Gecko/20100101 Firefox/88.0"}
        response=requests.get(url,params,headers=header)
        var=response.json()
        a=var['sessions']
        for i in a:
           
            if i['available_capacity']!=0:
                
                print(i["name"]+" "+i['vaccine'])
        
if __name__ == "__main__":
    
    #Input
    State=input("State name").lower()
    District=input("District name").lower()
    pc=input("Pincode")

    #function call
    State_Id=GetStateId(State)
    if State_Id==0:
        print("Invalid state \n")
    else:
        District_Id=GetDistrictId(District,str(State_Id))
        if District_Id==0:
             print("Invalid district \n")
        else:
            Availability(pc,District_Id)
            exit()
    valid=ValidPincode(pc)
    if valid==True:
        Availability(pc,District_Id)
    else:
        print("invalid pincode")
```
